```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

atos_res = np.random.binomial(1, 0.49, size = 1000 * 10**4).reshape(10**4, 1000)
atos_res = np.where(atos_res == 0, -1, 1)

def casino(atos):
    return atos.cumsum()


atos_res = np.apply_along_axis(casino, 1, atos_res)
atos_res += 1000
nom = np.random.randint(0, 10*4, size = 10)
atos_vis = atos_res[nom]

fig, ax = plt.subplots(figsize = (15, 5))

def vis(atos):
    ax.plot(np.arange(1, 1001), atos, color = '#CE1E83')
    return atos

atos = np.apply_along_axis(vis, 1, atos_vis[:len(atos_vis)])
ax.plot(np.arange(1, 1001), np.mean(atos_res, axis = 0), color = '#1B52C1', label = "Усредненная траектория")
ax.plot(np.arange(1, 1001), atos_vis[-1], color = '#CE1E83', label = "Эсперименты")
ax.set_xlabel("Номер посещения казино")
ax.set_ylabel("Богатство Атоса")
ax.set_title("Изменение бюджета Атоса после посещения казино", color = '#1B52C1', fontsize = 16)
plt.gca().set(xlim = (1, 1000))
plt.legend(fontsize = 12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

portos_res = np.zeros(10 ** 4) + 1000

def f_port(portos):
    for i in range(1000):
        if portos[-1] % 3 == 0:
            p = 0.09
        else:
            p = 0.74
        n = np.random.binomial(1, p) *2 - 1 
        portos = np.concatenate((portos, np.array([portos[-1]+n])), axis = 0)
    portos = portos[1::]
    return portos

portos_res = np.apply_along_axis(f_port, 1, portos_res[:, np.newaxis])
nom = np.random.randint(0, 10 ** 4, size = 10)
portos_vis = portos_res[nom]

fig, ax = plt.subplots(figsize = (15, 5))

def vis(portos):
    ax.plot(np.arange(1, 1001), portos, color = '#51E28B')
    return portos

atos = np.apply_along_axis(vis, 1, portos_vis[:len(portos_vis)])
ax.plot(np.arange(1, 1001), np.mean(portos_res, axis = 0), color = '#E1066D', label = "Усредненная траектория")
ax.plot(np.arange(1, 1001), portos_vis[-1], color = '#51E28B', label = "Эксперименты")
ax.set_xlabel("Номер посещения казино")
ax.set_ylabel("Бюджет Портоса")
ax.set_title("Изменение бюджета Портоса после посещения казино", color = '#E1066D', fontsize=16)
plt.gca().set(xlim=(1, 1000))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)


aramis_res = np.zeros(10**4) + 1000

def f_aramis(aram):
    p_s = np.random.binomial(1, 0.5, size = 1000)
    for p in p_s:
        if p == 1:
            money = np.random.binomial(1, 0.49) * 2 - 1
        else:
            if aram[-1] % 3 == 0:
                p1 = 0.09
            else:
                p1 = 0.74
            money = np.random.binomial(1, p1) * 2 - 1 
        aram = np.concatenate((aram, np.array([aram[-1] + money])), axis = 0)
    aram = aram[1::]
    return aram
    
    
aramis_res = np.apply_along_axis(f_aramis, 1, aramis_res[:, np.newaxis])
nom = np.random.randint(0, 10*4, size = 10)
aramis_vis = aramis_res[nom]

fig, ax = plt.subplots(figsize = (15, 5))

def vis(aram):
    ax.plot(np.arange(1, 1001), aram, color = '#046CA5')
    return aram

atos = np.apply_along_axis(vis, 1, aramis_vis[:len(aramis_vis)])
ax.plot(np.arange(1, 1001), np.mean(aramis_res, axis = 0), color = '#51E28B', label = "Усредненная траектория")
ax.plot(np.arange(1, 1001), aramis_vis[-1], color = '#046CA5', label = "Эксперименты")
ax.set_xlabel("Номер посещения казино")
ax.set_ylabel("Бюджет Арамиса")
ax.set_title("Изменение бюджета Арамиса после посещения казино", color = '#51E28B', fontsize=16)
plt.gca().set(xlim=(1, 1000))
plt.legend(fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

def f(n):
    x = np.where(scores > n, 1, 0)
    return sum(x)

res = np.random.normal(60, 10, size = 2 * 10000).reshape(10000, 2)
res_math = res[::, :1]
res_math = res_math.reshape(1, len(res_math))[0]
res_rus = res[::, -1:]
res_rus = res_rus.reshape(1, len(res_rus))[0]
scores = res.sum(1)
n = np.arange(80, 161, 5)
new_n = n[:, np.newaxis]    
win_for_n = np.apply_along_axis(f, 1, new_n)
win_for_n = win_for_n * 1000

fig, ax = plt.subplots(figsize = (10, 5))

ax.plot(n, win_for_n)

plt.title("Результаты отбора в УШЭ", color = 'b')
plt.xlabel("Проходной балл")
plt.ylabel("Число прошедших в УШЭ (тыс.)")
plt.style.use('bmh');
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

res_m = np.concatenate((res_math ** np.ones((17,1)), new_n), axis = 1)
res_r = np.concatenate((res_rus ** np.ones((17,1)), new_n), axis = 1)

def f_win(res):
    x = np.where(scores > res[-1], 1, 0)
    l = len(res)-1
    res = res[0:l]
    res = res * np.array(x)
    return res

  
    
math_win = np.apply_along_axis(f_win, 1, res_m)
rus_win = np.apply_along_axis(f_win, 1, res_r)

for math in math_win:
    math = math[math != 0]
for rus in rus_win:
    rus = rus[rus != 0]

a = np.sum(((math_win - np.mean(math_win)) * (rus_win - np.mean(rus_win))), axis = 1)
b = np.sum(((math_win - np.mean(math_win)) ** 2 ) * sum((rus_win - np.mean(rus_win)) ** 2), axis = 1)
b = b ** 0.5
cor = a / b

fig, ax = plt.subplots()
ax.plot(np.arange(80, 165, 5), cor, color = '#1B52C1', label = "Корреляция")
ax.set_xlabel("Проходной балл")
ax.set_ylabel("Коэффициент корреляции")
ax.set_title("Корреляция результатов среди не прошедших", color = '#1B52C1', fontsize = 16)
plt.legend(fontsize = 12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats # more style :)
from scipy.optimize import minimize
np.random.seed(1)

def lnL(theta, y):
    mu = theta[0]          
    s2 = np.exp(theta[1])  
    y = np.array(y)
    n = len(y)
    l = -0.5 * n * np.log(s2) - 0.5 * np.sum((y - mu)**2) / s2
    return -1 * l 

nif = np.random.normal(60, 10, size = 100)
lnL([60,100], nif)
res = minimize(lnL, [0, 0], args = nif)
res
```
